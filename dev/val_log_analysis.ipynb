```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast
```

```python
log_file = 'data/val_log.txt'

# load the dataframe (each line is a dictionary)
# Read the file and store each line as a dictionary
with open(log_file, 'r') as file:
    data = [eval(line.replace('inf', 'np.Inf')) for line in file]

# Creating DataFrame
df = pd.DataFrame(data)
```

```python
# add a column epoch to the dataframe
df['epoch'] = df.index + 1
```

```python
# plot recall vs epoch using seaborn
import seaborn as sns
sns.set_theme(style="darkgrid")
sns.lineplot(x="epoch", y="recall", data=df)
# plot precision vs epoch using seaborn
sns.lineplot(x="epoch", y="precision", data=df)
```

```python
# plot MAE_x vs epoch using seaborn
sns.lineplot(x="epoch", y="MAE_x", data=df)
```
